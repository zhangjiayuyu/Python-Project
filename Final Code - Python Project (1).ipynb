## 1.3 Data Slicing and Data Cleaning

### 1.3.1 The dataset contains data about candidates campaigning for 3 offices
#### 1. House of Representatives
#### 2. Senators
#### 3. Presidency

#### Since 78% of the dataset contains data pertaining to House of Representative election, we will focus on this category first

```python
#Create a new dataframe for only House of Representatives 

data = data[data['can_off'] == 'H'].reset_index(drop = True)
data.head()
```

```python
# Visualize missing data in this dataframe

msn.bar(data)
```

### 1.3.2 The dataset contains 52 columns with majority of them having missing data. 
### For our analysis purpose, we will be assuming that winning/losing is a function of
#### 1.Expenditure (net_ope_exp)
#### 2.Total Campaign Duration (column to be introduced)
#### 3.Party Affiliation (can_par_aff)
#### 4.State from which the candidate is running for that office (can_sta)
### As a result, we will be dropping the remaining columns. 

```python
# First let's create new column to capture total campaign duration

data['cov_sta_dat'] = pd.to_datetime(data['cov_sta_dat'])
data['cov_end_dat'] = pd.to_datetime(data['cov_end_dat'])
data['campaign_duration'] = (data['cov_end_dat'] - data['cov_sta_dat']).dt.days
data['can_off_dis'] = data['can_off_dis'].astype("str")
```

```python
# Create a new dataframe for columns which we are interested in.

data1 = data[['can_nam','can_off_sta','can_off_dis','can_par_aff', 'campaign_duration', 'net_ope_exp','winner']]
```

```python
data1.head()
```

```python
data1.shape
```

```python
# convert the net operating expenditure column to a float data type

value = data1["net_ope_exp"].str.replace('$','').str.replace(',','').str.replace('(','-').str.replace(')','').astype('float64')
data1["net_ope_exp"] = value
data1.dtypes
```

```python
# plot missing data in reduced dataset (data1)

msn.bar(data1)
```

```python
# replace missing values in winner table with N as they represent losers

data1['winner'] = data1['winner'].fillna('N')
data1.tail()
 
# dropping all the other rows that contains missing data
data1.dropna(how = 'any', inplace=True)
```

```python
#visualizing the bar again

msn.bar(data1)
```

# 3.1 Data Modelling - Logistic Regression
#### Since we are trying to predict whether a candidate will win or lose the election, this becomes a classification problem. Thus we will be using Logistic Regression to predict the outcome

```python
# import required libraries

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score, accuracy_score, log_loss
import sklearn.metrics as metrics
```

```python
#Assign dummy variables for 'can_off_dis','can_off_sta' and 'can_party_aff' using One Hot Encoding.

data1 = data1.drop(columns = ['can_nam'])
data1 = pd.concat((data1, pd.get_dummies(data1['can_off_dis']), pd.get_dummies(data1['can_off_sta']), pd.get_dummies(data1['can_par_aff'])), axis =1)

#Drop the original columns containing categorical variables
data1 = data1.drop(['can_off_sta'],axis =1).drop(['can_par_aff'],axis=1).drop(['can_off_dis'],axis=1)
```

```python
data1.head()
```

```python
#Assign Response and Predictor variables

X = data1.drop(['winner'], axis =1)
y = data1['winner'].map({'Y':1, 'N':0})
```

```python
#Split training and test data using Train Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 0)
```

```python
#Run Logistic Regression Model on train data

clf = LogisticRegression()

clf.fit(X_train, y_train)
```

```python
#Run the trained model on test data

y_pred = clf.predict(X_test)
```

```python
# Print the results

print('Trainig accuracy',clf.score(X_train,y_train)) 
print()
print('Testing accuracy',clf.score(X_test, y_test))
print()
print('Confusion Matrix')
print()
print('Recall:',recall_score(y_test,y_pred))
print()
print('Precision:',precision_score(y_test,y_pred))
print()
print('F1 Score:',f1_score(y_test,y_pred))
print()
print('Roc Auc Score:',roc_auc_score(y_test,y_pred))
```

```python
fpr, tpr, threshold = metrics.roc_curve(y_test, y_pred)
roc_auc = metrics.auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
